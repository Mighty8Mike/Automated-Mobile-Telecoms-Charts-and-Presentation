```python
# Step 1: Copy and clean columns (drop 20002007)
df_yrs = combined_df.copy()
df_yrs.columns = df_yrs.columns.astype(str)
years_to_drop = [str(y) for y in range(2000, 2008)]
df_yrs = df_yrs.drop(columns=years_to_drop)
print(f"Dropped columns: {years_to_drop}")
print(f"Remaining columns: {df_yrs.columns.tolist()}")
```

```python
# Step 2: Define helper function to normalize country names
def normalize_string(s):
    if isinstance(s, str):
        return unicodedata.normalize('NFKD', s).encode('ascii', 'ignore').decode('utf-8').strip().casefold()
    return s
```

```python
# Step 3: Normalize country names and find mismatches
subscribers = set(df_yrs.loc['Subscribers'].index.map(normalize_string))
arpu = set(df_yrs.loc['ARPU'].index.map(normalize_string))
population = set(df_yrs.loc['Population'].index.map(normalize_string))

all_countries = subscribers | arpu | population
mismatched_normalized = [
    c for c in all_countries
    if not (c in subscribers and c in arpu and c in population)
]
print(f"\n Countries not in all 3 categories (normalized):")
for c in sorted(mismatched_normalized): print(" -", c)
print(f" Total mismatched (normalized): {len(mismatched_normalized)}")
```

```python
# Step 4: Drop rows with mismatched countries
df_yrs = df_yrs[~df_yrs.index.get_level_values(1).map(normalize_string).isin(mismatched_normalized)]
print(f"\n Removed mismatched countries. Remaining rows: {df_yrs.shape[0]}")
```

```python
# Step 5: Validate by counting appearances per country
counts = df_yrs.index.get_level_values(1).value_counts()
incomplete = counts[counts != 3].index.tolist()
if incomplete:
    print("\nStill incomplete countries (check for subtle naming differences):")
    for country in sorted(incomplete): print(" -", country)
else:
    print("\nAll countries now appear in exactly 3 categories (Subscribers, ARPU, Population).")
```

```python
# If we really want to clean most zeros
# Ensure year columns are strings
df_yrs.columns = df_yrs.columns.astype(str)

# Select year columns except 2024
year_columns_excl_2024 = [col for col in df_yrs.columns if col != '2024']

# Identify countries to drop
countries_to_drop = []

for country in df_yrs.index.get_level_values(1).unique():
    try:
        subset = df_yrs.xs(country, level=1)  # All 3 metrics for this country
        zero_counts = (subset[year_columns_excl_2024] == 0).sum(axis=1)
        if (zero_counts >= 2).any():
            countries_to_drop.append(country)
    except KeyError:
        continue

# Drop all rows for those countries
df_yrs = df_yrs[~df_yrs.index.get_level_values(1).isin(countries_to_drop)]

#  Summary
print(f" Dropped all rows for {len(countries_to_drop)} countries with 2+ zeros (excluding 2024).")
print("Dropped countries:", countries_to_drop)
```

```python
required_metrics = {'Population', 'Subscribers', 'ARPU'}
cols_ex_2024 = [col for col in df_yrs.columns if col != '2024']

to_drop = []

for country in df_yrs.index.get_level_values(1).unique():
    present_metrics = set(df_yrs.loc[(slice(None), country), :].index.get_level_values(0))
    if not required_metrics.issubset(present_metrics):
        to_drop.append(country)
        continue

    subset = df_yrs.xs(country, level=1)
    zero_counts = (subset[cols_ex_2024].fillna(0) == 0).sum(axis=1)
    if (zero_counts >= 2).any():
        to_drop.append(country)

df_yrs = df_yrs[~df_yrs.index.get_level_values(1).isin(to_drop)]

print(f"Dropped {len(to_drop)} countries with missing metrics or 2 zeros (excluding 2024):", to_drop)
```

```python
# Add Level-1 Blocks Market Size and Penetration Rate 

year_cols = [str(y) for y in range(2008, 2025)]

# 1) grab clean DataFrames for each Indicator
subs = df_yrs.xs('Subscribers', level='Indicator')
arpu = df_yrs.xs('ARPU', level='Indicator')
pop  = df_yrs.xs('Population', level='Indicator')

# 2) reset indices so meta-columns become ordinary columns
subs_reset = subs.reset_index()
arpu_reset = arpu.reset_index()
pop_reset  = pop.reset_index()

# 3) calculate market size
market_size = subs_reset[year_cols].astype(float).values * arpu_reset[year_cols].astype(float).values * 12   # annualized ARPU
market_size_df = pd.DataFrame(market_size, columns=year_cols)
for col in ['entityName', 'ITU Region', 'World Bank Income group (2024 July)']:
    market_size_df[col] = subs_reset[col]
market_size_df['dataYear'] = 'Market Size'
market_size_df = market_size_df[['dataYear','entityName','ITU Region','World Bank Income group (2024 July)'] + year_cols]

# 4) calculate penetration rate
penetration = subs_reset[year_cols].astype(float).values / pop_reset[year_cols].astype(float).replace(0, np.nan).values
penetration_df = pd.DataFrame(penetration, columns=year_cols)
for col in ['entityName', 'ITU Region', 'World Bank Income group (2024 July)']:
    penetration_df[col] = subs_reset[col]
penetration_df['dataYear'] = 'Penetration Rate'
penetration_df = penetration_df[['dataYear','entityName','ITU Region','World Bank Income group (2024 July)'] + year_cols]

# 5) stitch back into your big df
combined = pd.concat([market_size_df, penetration_df], ignore_index=True)
combined = combined.set_index(['dataYear','entityName','ITU Region','World Bank Income group (2024 July)'])
df_indicators = pd.concat([df_yrs, combined])

```

```python
#   1. Averages of ARPU and Penetration Rate for Level-3 and Level-4 (leave Level-2 empty)
#   2. Totals of Population, Subscribers, Market Size for Level-3 and Level-4 (leave Level-2 empty)

# Step 1: Select numeric columns (i.e., year columns)
numeric_cols = year_cols  # e.g., ['2018', '2019', ..., '2023']

# Step 2: Define aggregation categories
avg_years = ['ARPU', 'Penetration Rate']
sum_years = ['Population', 'Subscribers', 'Market Size']

# Step 3: Create containers
avg_dfs = []
sum_dfs = []

# Step 4: Group and aggregate averages (ARPU, Penetration Rate)
for indicator in avg_years:
    if indicator in df_yrs.index.get_level_values(0):
        df_filtered = df_yrs.loc[indicator]
        df_avg = df_filtered.groupby(['ITU Region', 'World Bank Income group (2024 July)'])[numeric_cols].mean()

        # Add back the full MultiIndex structure
        df_avg['dataYear'] = indicator
        df_avg['entityName'] = 'Average - Region & Income Group'
        df_avg = df_avg.reset_index()
        df_avg = df_avg[['dataYear', 'entityName', 'ITU Region', 'World Bank Income group (2024 July)'] + numeric_cols]
        df_avg = df_avg.set_index(['dataYear', 'entityName', 'ITU Region', 'World Bank Income group (2024 July)'])

        avg_dfs.append(df_avg)

# Step 5: Group and aggregate totals (Population, Subscribers, Market Size)
for indicator in sum_years:
    if indicator in df_yrs.index.get_level_values(0):
        df_filtered = df_yrs.loc[indicator]
        df_sum = df_filtered.groupby(['ITU Region', 'World Bank Income group (2024 July)'])[numeric_cols].sum()

        # Add back the full MultiIndex structure
        df_sum['dataYear'] = indicator
        df_sum['entityName'] = 'Total - Region & Income Group'
        df_sum = df_sum.reset_index()
        df_sum = df_sum[['dataYear', 'entityName', 'ITU Region', 'World Bank Income group (2024 July)'] + numeric_cols]
        df_sum = df_sum.set_index(['dataYear', 'entityName', 'ITU Region', 'World Bank Income group (2024 July)'])

        sum_dfs.append(df_sum)

# Step 6: Combine all aggregated frames
aggregated_df = pd.concat(avg_dfs + sum_dfs)

# Step 7: Append to df_indicators
df_totavg = df_indicators.copy() 
df_totavg = pd.concat([df_totavg, aggregated_df])
```

```python
# Clean the data: replace inf and fill NA with 0
df_clean = df_totavg.replace([np.inf, -np.inf], np.nan).fillna(0)

# Identify 'Penetration Rate' rows using MultiIndex level
mask_pr = df_clean.index.get_level_values(0) == 'Penetration Rate'

# Convert all data to float (coerce errors), fill NA with 0
df_fmt = df_clean.copy().apply(pd.to_numeric, errors='coerce').fillna(0).astype(float)  #  ENSURE FLOAT

# Create empty display DataFrame with same structure
df_display = pd.DataFrame(index=df_fmt.index, columns=df_fmt.columns)

# Format each value for display
for idx in df_fmt.index:
    for col in df_fmt.columns:
        val = df_fmt.at[idx, col]
        
        if idx[0] == 'Penetration Rate':
            #  Leave as float, show % with 4 decimal precision
            df_display.at[idx, col] = f"{val:.4f}%"  # e.g. '92.4321%'
        else:
            #  Leave float unrounded, format to 2 decimal places
            df_display.at[idx, col] = f"{val:.2f}"   # e.g. '16.63'

# Copy final formatted values to df_fmt for downstream use
df_fmt = df_display.copy()

# Restore index/column names (important for Excel/export)
df_fmt.index.set_names(['Key Indicator', 'Country', 'ITU Region', 'WB Income Group'], inplace=True)
df_fmt.columns.name = 'Year'

```
